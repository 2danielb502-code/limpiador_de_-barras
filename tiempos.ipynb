```python
import cv2
import numpy as np
import random
import os

width = 3000
height = 3000
batch_size = 20   # 20 imágenes

def generar_imagen():

    canvas = np.ones((height, width, 3), dtype=np.uint8) * 255

    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 5
    thickness = 10
    color = (0, 0, 0)

    cols = 10
    rows = 10

    cell_width = width // cols
    cell_height = height // rows

    for row in range(rows):
        for col in range(cols):

            numero = str(random.randint(0, 9))

            x_center = col * cell_width + cell_width // 2
            y_center = row * cell_height + cell_height // 2

            (text_width, text_height), _ = cv2.getTextSize(
                numero, font, font_scale, thickness
            )

            x = x_center - text_width // 2
            y = y_center + text_height // 2

            cv2.putText(canvas, numero, (x, y),
                        font, font_scale,
                        color, thickness,
                        cv2.LINE_AA)

    return canvas


# Ruta Descargas
downloads_path = os.path.join(os.path.expanduser("~"), "Downloads")

# Generar y guardar imágenes
for i in range(batch_size):
    img = generar_imagen()
    filename = os.path.join(downloads_path, f"imagen_{i+1}.png")
    cv2.imwrite(filename, img)

print("Listo: 20 imágenes generadas en Descargas")
```
