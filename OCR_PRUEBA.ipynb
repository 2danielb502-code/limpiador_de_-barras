```python
import cv2
import os
from easyocr import Reader


# CONFIGURACIÓN DE CARPETAS


input_folder = r"C:\Users\dbecerraq\Documents\Prueba_2"
output_image_folder = r"C:\Users\dbecerraq\Documents\PROCESO COMPLETO\prueba_3"
output_txt_folder = r"C:\Users\dbecerraq\Documents\PROCESO COMPLETO\Archivos_txt"

os.makedirs(output_image_folder, exist_ok=True)
os.makedirs(output_txt_folder, exist_ok=True)

reader = Reader(["en"], gpu=False)


# PROCESAMIENTO AUTOMÁTICO


for filename in os.listdir(input_folder):

    if filename.lower().endswith((".png", ".jpg", ".jpeg")):

        print(f"\nProcesando: {filename}")

        image_path = os.path.join(input_folder, filename)

        image = cv2.imread(image_path, 0)  # gris
        image_color = cv2.imread(image_path, 1)  # color

        h_image, w_image = image.shape[:2]

        image_boxes = reader.readtext(image, detail=1, allowlist="0123456789")

        image_color_copy = image_color.copy()

        txt_filename = os.path.splitext(filename)[0] + ".txt"
        txt_path = os.path.join(output_txt_folder, txt_filename)

        with open(txt_path, "w") as txt_file:

            red_color = 50

            for image_box in image_boxes:

                print("codigo:", image_box[1])

                # Coordenadas
                xi, yi = list(map(int, image_box[0][0]))
                xf, yf = list(map(int, image_box[0][2]))

                print("Coordenadas:", xi, yi, xf, yf)

                # Centro
                cx = (xi + xf) // 2
                cy = (yi + yf) // 2

                # Dibujar rectángulo
                cv2.rectangle(image_color_copy,
                              (xi, yi),
                              (xf, yf),
                              (red_color, 0, 0),
                              2)

                # Dibujar centro
                cv2.circle(image_color_copy,
                           (cx, cy),
                           3,
                           (0, 200, 0),
                           -1)

                red_color += 50

                # Ancho y alto
                w = xf - xi
                h = yf - yi

                print("Ancho y alto:", w, h)

                # Normalización (como tú lo hacías)
                cx_norm = round((cx / w_image), 6)
                cy_norm = round((cy / h_image), 6)
                w_norm = round((w / w_image), 6)
                h_norm = round((h / h_image), 6)

                code_class = image_box[1]

                print("Normalizado:",
                      code_class,
                      cx_norm,
                      cy_norm,
                      w_norm,
                      h_norm)

                # Guardar en TXT
                txt_file.write(
                    f"{code_class} {cx_norm} {cy_norm} {w_norm} {h_norm}\n"
                )

        # Guardar imagen procesada
        output_image_path = os.path.join(output_image_folder, filename)
        cv2.imwrite(output_image_path, image_color_copy)

        print("Guardado imagen y txt correctamente.")

        del image
        del image_color
        del image_color_copy

print("\n✅ PROCESO MASIVO TERMINADO.")
```
