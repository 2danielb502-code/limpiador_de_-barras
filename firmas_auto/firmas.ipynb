```python
from openpyxl import load_workbook
from openpyxl.drawing.image import Image
import os
import win32com.client as win32 

#carpeta de entrada 
carpeta_entrada = r"C:\Users\dbecerraq\Documents\carpeta_de_entrada"

#carpeta de salida
carpeta_de_salida = r"C:\Users\dbecerraq\Documents\carpeta_de_salida" 

# imagen de firma 
firma_path = r"C:\Users\dbecerraq\Documents\prueba_excel\firma.jpg"

# Excel COM 
excel = win32.Dispatch("Excel.Application")
excel.Visible = False

#corredor y buscador de ecxeles
for archivo in os.listdir(carpeta_entrada):
    if archivo.endswith(".xlsx"):

        ruta_archivo = os.path.join(carpeta_entrada, archivo)
        print(f"prosesando: {archivo}")

        wb = load_workbook(ruta_archivo)
        hoja = wb.active

        firma = Image(firma_path)
        firma.width = 150 
        firma.height = 60

        texto_busqueda = "firma de la persona con rol de jefe inmediato"

        celda_encontrada = None

        # ancla de firmas
        for row in hoja.iter_rows():
            for cell in row:
                if cell.value:
                    if str(cell.value).strip().lower() == texto_busqueda:
                        celda_encontrada = cell
                        break
            if celda_encontrada:
                break

        # salto de celda
        if celda_encontrada:
            fila_arriba = celda_encontrada.row - 3
            columna = celda_encontrada.column_letter
            destino = f"{columna}{fila_arriba}"

            hoja.add_image(firma, destino)
            print(f"firma incertada en {destino}")

            # ajuste de centrado de imagen
            hoja.row_dimensions[fila_arriba].height = 5
            hoja.row_dimensions[fila_arriba + 1].height = 40

        else:
            print("No se encontro el ancla")

        # guardado de salida en carpeta 
        salida = os.path.join(carpeta_de_salida, f"firmado_{archivo}")
        wb.save(salida)

        #convertir a pdf
        pdf_salida = salida.replace(".xlsx", ".pdf")

        #adicionales
        libro = excel.Workbooks.Open(salida)
        libro.ExportAsFixedFormat(0, pdf_salida)
        libro.Close()

        print(f"pdf generado: {pdf_salida}")

# cerrar excel
excel.Quit()

print("firma ya guardada")
```
