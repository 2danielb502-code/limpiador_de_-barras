```python
from pdf2image import convert_from_path
import cv2
import matplotlib.pyplot as plt
import os

os.environ["PATH"] += r";C:\Users\dbecerraq\Documents\DANIEL_BECERRA\PROGRAMAS_INSTALADOS\poppler-0.67.0\bin"

# Carpeta de entrada
pdf_folder = r"C:\Users\dbecerraq\Documents\DANIEL_BECERRA\PROYECTOS\E-14\utilities\PDF"

# Carpeta de salida
base_output = r"C:\Users\dbecerraq\Documents\DANIEL_BECERRA\PROYECTOS\E-14\utilities\results_interpreter_basic"

# Recorrer todos 
for file in os.listdir(pdf_folder):
    if file.endswith(".pdf"):

        pdf_path = os.path.join(pdf_folder, file)
        name_pdf = os.path.splitext(file)[0]

        print(f"Procesando: {name_pdf}")

        # Crear carpeta para cada PDF
        output_folder = os.path.join(base_output, name_pdf)
        os.makedirs(output_folder, exist_ok=True)

        # PDF a imágenes
        images = convert_from_path(
            pdf_path,
            dpi=300,
            # poppler_path=r"C:\poppler\Library\bin"
        )

        # Guardar imágenes dentro de su carpeta original
        for i, img in enumerate(images):
            img.save(
                os.path.join(output_folder, f"{name_pdf}_pagina_{i+1}.png"),
                "PNG"
            )
```

```python
from pdf2image import convert_from_path
import cv2
import matplotlib.pyplot as plt
import os

# Agregar Poppler al PATH
os.environ["PATH"] += r";C:\Users\dbecerraq\Documents\DANIEL_BECERRA\PROGRAMAS_INSTALADOS\poppler-0.67.0\bin"

# Carpeta de entrada (PDFs)
pdf_folder = r"C:\Users\dbecerraq\Documents\ejemplo_muestra e-14\PDF"

# Carpeta de salida
base_output = r"C:\Users\dbecerraq\Documents\ejemplo_muestra e-14\salida_uno"

# Recorrer todos los archivos en la carpeta
for file in os.listdir(pdf_folder):

    # Solo archivos PDF
    if file.lower().endswith(".pdf"):

        pdf_path = os.path.join(pdf_folder, file)
        name_pdf = os.path.splitext(file)[0]

        print(f"Procesando: {name_pdf}")

        # Crear carpeta para cada PDF
        output_folder = os.path.join(base_output, name_pdf)
        os.makedirs(output_folder, exist_ok=True)

        # Convertir PDF a imágenes
        images = convert_from_path(
            pdf_path,
            dpi=300,
            poppler_path=r"C:\Users\dbecerraq\Documents\DANIEL_BECERRA\PROGRAMAS_INSTALADOS\poppler-0.67.0\bin"
        )

        # Guardar imágenes en su carpeta correspondiente
        for i, img in enumerate(images):
            img.save(
                os.path.join(output_folder, f"{name_pdf}_pagina_{i+1}.png"),
                "PNG"
            )

        print(f"✔ Terminado: {name_pdf}\n")
```
