```python
import cv2
import os
import zxingcpp

# 🔹 Ruta de la carpeta (CAMBIA ESTO SI ES NECESARIO)
carpeta = r"C:\Users\dbecerraq\Documents\Daniel\Daniel\images"

# 🔹 Verificar que la ruta exista
if not os.path.exists(carpeta):
    print("❌ La ruta no existe. Verifica la carpeta.")
    exit()

# 🔹 Función de preprocesamiento
def preprocesar(imagen):
    gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

    # Aumentar contraste
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    gris = clahe.apply(gris)

    # Reducir ruido
    gris = cv2.GaussianBlur(gris, (3, 3), 0)

    return gris

# 🔹 Procesar carpeta
def procesar_carpeta(ruta_carpeta):
    archivos = os.listdir(ruta_carpeta)

    if not archivos:
        print("⚠️ La carpeta está vacía")
        return

    for archivo in archivos:
        ruta_imagen = os.path.join(ruta_carpeta, archivo)

        try:
            imagen = cv2.imread(ruta_imagen)

            if imagen is None:
                continue

            print(f"\n📂 Procesando: {archivo}")

            # 🔹 Preprocesar imagen
            img_proc = preprocesar(imagen)

            resultados = []

            # 🔹 Intento 1: imagen procesada
            resultados = zxingcpp.read_barcodes(img_proc)

            # 🔹 Intento 2: imagen original
            if not resultados:
                resultados = zxingcpp.read_barcodes(imagen)

            # 🔹 Intento 3: diferentes escalas (muy útil)
            if not resultados:
                for escala in [1.5, 2.0, 3.0]:
                    resized = cv2.resize(imagen, None, fx=escala, fy=escala)
                    resultados = zxingcpp.read_barcodes(resized)
                    if resultados:
                        print(f"🔍 Detectado con escala x{escala}")
                        break

            # 🔹 Resultados
            if not resultados:
                print("❌ No se detectó ningún código")
            else:
                for r in resultados:
                    print("✅ Formato:", r.format)
                    print("📦 Datos:", r.text)
                    print("----------------------")

        except Exception as e:
            print(f"⚠️ Error procesando {archivo}: {e}")

# 🔹 Ejecutar
procesar_carpeta(carpeta)
```
